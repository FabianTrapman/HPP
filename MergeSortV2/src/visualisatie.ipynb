# MergeSort

```python
def run_cpp_program():
    result = subprocess.run(["./my_cpp_program.exe"], stdout=subprocess.PIPE, text=True)
    return result.stdout
```

Hier visualiseer ik mijn threaded MergeSort algoritme met een run-time complexiteits analyse

Om te demonstreren dat mijn algoritme naar behoren werkt doen doen we als eerst een test run met 64 elementen en 8 threads

# 8 Threads

```python
def analyze_timing():
    output = run_cpp_program()
    timing_values = [tuple(map(int, pair.split(', '))) for pair in output.split('\n') if pair]

    print(timing_values)

    # Unpack the tuples into separate lists for x and y values
    x_values, y_values = zip(*timing_values)
    x_values = list(map(str, x_values))

    # Plot the data
    plt.bar(x_values, y_values, color='blue')
    plt.title('run-time complexiteit analyse')
    plt.xlabel('Threads')
    plt.ylabel('Tijd')
    plt.show()
```

```python
analyze_timing()
```
